# First Experiment

Here, we use `add_two_numbers`, make a nice plot and store it in the `results` folder.

## 1. Run experiment

```python
from our_library.some_functions import add_10
```

```python
# Let's use our function `add_10`
numbers = [1, 2, 3, 4, 5, 8]
numbers_add_10 = [add_10(a) for a in numbers]
print("numbers_add_10 =", numbers_add_10)
```

## 2. Save results

```python
import os
from our_library.paths import get_results_subdir, save_json
```

```python
# Let's save the results
results = {"numbers": numbers, "numbers_add_10": numbers_add_10}
savedir = get_results_subdir("01_first_experiment")
save_json(results, os.path.join(savedir, "results"))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = ["svg"]
```

```python
# Let's make a nice plot
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
ax.plot(numbers, numbers_add_10)
ax.set_xlabel("a")
ax.set_ylabel("add_10(a)")
ax.set_title("Visualization exp01")
fig.tight_layout()
plt.show()
fig.savefig(os.path.join(savedir, "plot.pdf"))
```
